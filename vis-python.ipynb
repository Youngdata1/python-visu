```python
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv ('Video games.csv', sep = ',')
dataset

```

```python
#Cleaning the data
dataset.dropna()
```

```python
#Creating an area chart to show the trend of global sales over the years
import matplotlib.pyplot as plt
plt.fill_between(dataset['Year'], dataset['Global_Sales'])
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Trend of Global Sales over the Years')
plt.show()
```

```python
#Creating a histogram for global sales
plt.hist(dataset['Global_Sales'], bins=10)
plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Global Sales')
plt.show()
```

```python
# Creating a scatter plot for North American sales and European sales
plt.scatter(dataset['NA_Sales'], dataset['EU_Sales'])
plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.title('Relationship between North American Sales and European Sales')
plt.show()
```

```python
# Creating a scatter plot for Japanese sales and Other sales
plt.scatter(dataset['JP_Sales'], dataset['Other_Sales'])
plt.xlabel('Japanese Sales')
plt.ylabel('Other Sales')
plt.title('Relationship between Japanese Sales and Other Sales')
plt.show()
```

```python
# Grouping the data by platform and year and calculating the total sales for each combination
platform_sales = dataset.groupby(['Platform', 'Year'])['Global_Sales'].sum().unstack()

# Creating a stacked area chart for the distribution of sales by platform over the years
platform_sales.plot.area()
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Distribution of Sales by Platform over the Years')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Calculate the total global sales for each publisher and select the top 10
top_publishers = dataset.groupby('Publisher')['Global_Sales'].sum().nlargest(10)

# Create a bar chart for the top 10 publishers by global sales
plt.bar(top_publishers.index, top_publishers.values)
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Top 10 Publishers by Global Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculating the total global sales for each genre
genre_sales = dataset.groupby('Genre')['Global_Sales'].sum()

# Create a pie chart for the distribution of genres by global sales
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%')
plt.title('Distribution of Genres by Global Sales')
plt.axis('equal')
plt.show()
```
